```python
import requests
import pandas as pd

API_URL = "https://data.melbourne.vic.gov.au/api/explore/v2.1/catalog/datasets/soil-sensor-readings-historical-data/records"
BATCH_SIZE = 100
TOTAL_RECORDS = 200  # Adjust as needed

def fetch_data(limit, offset):
    params = {
        "limit": limit,
        "offset": offset,
        "order_by": "local_time desc",
        "timezone": "Australia/Sydney"
    }
    response = requests.get(API_URL, params=params)
    if response.status_code == 200:
        data = response.json()["results"]
        return data
    else:
        print(f"Failed to fetch data at offset {offset}")
        return []

def main():
    all_records = []
    for offset in range(0, TOTAL_RECORDS, BATCH_SIZE):
        batch = fetch_data(BATCH_SIZE, offset)
        all_records.extend(batch)

    # Convert to DataFrame
    df = pd.DataFrame(all_records)

    # Save to CSV
    df.to_csv("soil_sensor_data.csv", index=False)
    print(f"Saved {len(df)} records to soil_sensor_data.csv")

if __name__ == "__main__":
    main()
```
